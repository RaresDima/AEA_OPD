```python
stats = {}
for dataset in results:
    stats[dataset] = {}
    for configuration in results[dataset]:
        if "_old" in dataset:
            dataset_alternative = dataset.replace("old","improved")
        if "_improved" in dataset:
            dataset_alternative = dataset.replace("improved","old")
            
        mean_time = mean(results[dataset][configuration]["times_taken"])
        max_time = max(results[dataset][configuration]["times_taken"])
        min_time = min(results[dataset][configuration]["times_taken"])
        min_time = min(results[dataset][configuration]["times_taken"])
        std_time = tstd(results[dataset][configuration]["times_taken"])
        
        fitness_histories = []
        timeouts = []
        for run in results[dataset][configuration]["results"]:
            fitness_histories.append(run["fitness_history"])
            timeouts.append(run["timeout"])
            
        fitness_histories_alternative = []
        timeouts_alternative = []
        for run in results[dataset_alternative][configuration]["results"]:
            fitness_histories_alternative.append(run["fitness_history"])
            timeouts_alternative.append(run["timeout"])
            
        mean_fitness = [np.array(fitness_history).mean() for fitness_history in fitness_histories]
        max_fitness = [np.array(fitness_history).max() for fitness_history in fitness_histories]
        min_fitness = [np.array(fitness_history).min() for fitness_history in fitness_histories]
        
        min_fitness_alternative = [np.array(fitness_history).min() for fitness_history in fitness_histories_alternative]
        
        mean_mean_fitness = np.array(mean_fitness).mean()
        mean_max_fitness = np.array(max_fitness).mean()
        mean_min_fitness = np.array(min_fitness).mean()
        std_min_fitness = np.array(min_fitness).std()
        
        mean_timeout = np.array(timeouts, dtype=int).mean()
        
        
        conf_int = sem(results[dataset][configuration]["times_taken"]) * t.ppf((1 + 0.95) / 2, len(results[dataset][configuration]["times_taken"]) - 1)

        start_mean_time_conf_int = mean_time - conf_int
        end_mean_time_conf_int = mean_time+ conf_int
        
        
        conf_int = sem(min_fitness) * t.ppf((1 + 0.95) / 2, len(min_fitness) - 1)

        start_min_fitness_conf_int = mean_min_fitness - conf_int
        end_min_fitness_conf_int = mean_min_fitness+ conf_int
        
        time_t_statistic,time_pvalue = ttest_ind(results[dataset][configuration]["times_taken"],results[dataset_alternative][configuration]["times_taken"], equal_var=False, nan_policy='omit')
        min_fitness_t_statistic,min_fitness_pvalue = ttest_ind(min_fitness,min_fitness_alternative, equal_var=False, nan_policy='omit')
        
        
        
        stats[dataset][configuration] = {'mean_time':mean_time, 'max_time':max_time, 'min_time':min_time,'std_time':std_time,
                                         'mean_mean_fitness':mean_mean_fitness, 'mean_max_fitness':mean_max_fitness, 'mean_min_fitness':mean_min_fitness,'std_min_fitness':std_min_fitness,
                                        'mean_timeout':mean_timeout,
                                        'start_mean_time_conf_int':start_mean_time_conf_int,
                                        'end_mean_time_conf_int':end_mean_time_conf_int,
                                        'start_min_fitness_conf_int':start_min_fitness_conf_int,
                                        'end_min_fitness_conf_int':end_min_fitness_conf_int,
                                        'Standard error of the time mean':sem(results[dataset][configuration]["times_taken"]),
                                        'Standard error of the fitness mean':sem(min_fitness),
                                        'time_t_statistic':time_t_statistic,
                                        'time_pvalue':time_pvalue,
                                        'min_fitness_t_statistic':min_fitness_t_statistic,
                                        'min_fitness_pvalue':min_fitness_pvalue}
```

```python
headers = ['dataset','configuration','mean_time','max_time','min_time','std_time','mean_mean_fitness','mean_max_fitness','mean_min_fitness','std_min_fitness','mean_timeout','start_mean_time_conf_int','end_mean_time_conf_int','start_min_fitness_conf_int','end_min_fitness_conf_int','Standard error of the time mean','Standard error of the fitness mean','time_t_statistic','time_pvalue','min_fitness_t_statistic','min_fitness_pvalue']
```

```python
table = []
for dataset in stats:
    for configuration in stats[dataset]:
        table.append((dataset,configuration) + tuple(stats[dataset][configuration].values()))
```

```python
display(HTML(tabulate.tabulate(table, tablefmt='html', headers = headers).replace('nan','0')))
```
